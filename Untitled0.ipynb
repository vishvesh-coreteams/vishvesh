```python
# adding feature selection

# Assuming your dataset has columns: 'Product_ID', 'Month', 'Quantity_Sold'
# df = pd.read_csv("your_data.csv")

# Convert 'Month' to datetime format
df['year_month'] = pd.to_datetime(df['year_month'])

# Sort data by Product_ID and Month
df = df.sort_values(by=['product_id', 'year_month'])

# Create a new column for last month's sales
df['Lag_1'] = df.groupby('product_id')['quantity_sold'].shift(1)

# Create lag for 2 months ago
df['Lag_2'] = df.groupby('product_id')['quantity_sold'].shift(2)
```

```python
# 3-month moving average
df['Rolling_Mean_3'] = df.groupby('product_id')['quantity_sold'].transform(lambda x: x.rolling(3, min_periods=1).mean())

# 6-month moving average
df['Rolling_Mean_6'] = df.groupby('product_id')['quantity_sold'].transform(lambda x: x.rolling(6, min_periods=1).mean())
```

```python
# Extract month number from date
df['Month_Num'] = df['year_month'].dt.month

# Encode cyclical month feature (useful for models like XGBoost)
df['Month_Sin'] = np.sin(2 * np.pi * df['Month_Num'] / 12)
df['Month_Cos'] = np.cos(2 * np.pi * df['Month_Num'] / 12)
```

```python
# Example: Adding a column for flu season (assuming flu season is from November to March)
df['Flu_Season'] = df['Month_Num'].apply(lambda x: 1 if x in [11, 12, 1, 2, 3] else 0)
```

```python
df.fillna(0, inplace=True)  # Replace NaNs with 0 (or use forward-fill method)
```

```python
# Implement Holt-Winters Exponential Smoothing

from statsmodels.tsa.holtwinters import ExponentialSmoothing


from google.colab import files
uploaded = files.upload()

# Load dataset
df = pd.read_csv("anahadSales.csv")
df['year_month'] = pd.to_datetime(df['year_month'])
df = df.sort_values(by=['product_id', 'year_month'])

# Select one product to test first
product_id = 'some_product'
product_df = df[df['product_id'] == product_id].set_index('month')

# Train model
model = ExponentialSmoothing(
    product_df['quantity_sold'],
    seasonal='add',
    seasonal_periods=12
)
model_fit = model.fit()

# Forecast for next 6 months
future_steps = 6
forecast = model_fit.forecast(steps=future_steps)

# Plot results
plt.figure(figsize=(10,5))
plt.plot(product_df.index, product_df['quantity_sold'], label='Actual Sales')
plt.plot(forecast.index, forecast, label='Forecast', linestyle='dashed', color='red')
plt.legend()
plt.title(f"Forecast for Product {product_id}")
plt.show()
```
